```python
from datasets import load_dataset
import pandas as pd
```

```python
PROMPT_TEMPLATE = """Here is a forecasting question: {question}\n\nHere are the resolution criteria of the question: {resolution_criteria}.\n\nBased on the forecasting question and resolution criteria, generate {quality_type} quality step-by-step reasoning to create a resolution. Output exactly 6 steps without an introduction or a conclusion."""
print(PROMPT_TEMPLATE)
```

```python
dataset = load_dataset(path="YuehHanChen/forecasting", split="train")
dataset
```

```python
df_train = pd.DataFrame(dataset)
df_train
```

```python
df_train.loc[0, 'question']
```

```python
def format_prompt(df: pd.DataFrame, row: int, is_high_quality: bool) -> str:
    assert isinstance(is_high_quality, bool)
    return PROMPT_TEMPLATE.format(question = df.loc[row, "question"],
                                  resolution_criteria = df.loc[row, "resolution_criteria"],
                                  quality_type = "high" if is_high_quality else "low")
```

```python
print(format_prompt(df_train, 0, True))
```

```python
print(format_prompt(df_train, 0, False))
```
